```python
from google.cloud import texttospeech
from pydub import AudioSegment
from IPython.display import Audio, display
from pydub.playback import play
import pandas as pd
import io
import os
```

```python
df = pd.read_csv('/content/drive/MyDrive/PhD/data/target_sentences.csv')
df
```

```python
unique_words_list = []
# Process each row in the DataFrame
for sentence in df['Text']:
    words = sentence.lower().split()
    for word in words:
      unique_words_list.append(word)

unique_words_list = set(unique_words_list)
```

```python
len(unique_words_list)
```

```python
# Replace 'your-file-name.json' with the name of your uploaded file
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/sincere-kit-430523-n8-70d333752e6c.json"

# Initialize the Text-to-Speech client
client = texttospeech.TextToSpeechClient()

voice = texttospeech.VoiceSelectionParams(language_code="sr-RS", ssml_gender=texttospeech.SsmlVoiceGender.NEUTRAL)
audio_config = texttospeech.AudioConfig(audio_encoding=texttospeech.AudioEncoding.LINEAR16)
```

```python
def synthesize_word(word, filename):
    """Synthesize a single word and save to a file."""
    input_text = texttospeech.SynthesisInput(text=word)
    response = client.synthesize_speech(input=input_text, voice=voice, audio_config=audio_config)
    with open(filename, "wb") as out:
        out.write(response.audio_content)
        print(f'Audio content for "{word}" written to file "{filename}"')

# List of words to synthesize
filenames = []

# Synthesize each word
for word in unique_words_list:
    filename = f"/content/drive/MyDrive/PhD/Generated Words/{word}.wav"
    filenames.append(filename)
    synthesize_word(word, filename)
```
